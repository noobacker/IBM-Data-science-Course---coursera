```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your own data)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015],
    'Sedan': [500, 600, 750, 800, 700, 900],
    'SUV': [300, 400, 450, 550, 600, 700],
    'Truck': [200, 250, 300, 350, 400, 450]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Create a line chart with different lines for vehicle types
plt.figure(figsize=(10, 6))

# Plot lines for each vehicle type
for column in df.columns[1:]:
    plt.plot(df['Year'], df[column], marker='o', label=column)

plt.title('Automobile Sales by Vehicle Type Over the Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt
```
