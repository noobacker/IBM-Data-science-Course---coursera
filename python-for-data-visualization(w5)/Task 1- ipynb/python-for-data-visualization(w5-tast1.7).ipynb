```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
expenditure = [500000, 300000]  # Replace with the actual values for recession and non-recession periods

# Labels for the pie chart
labels = ['Recession', 'Non-Recession']

# Explode a slice if required (e.g., 'Recession')
explode = (0.1, 0)

# Create the pie chart
plt.pie(expenditure, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Advertising Expenditure during Recession and Non-Recession Periods')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
