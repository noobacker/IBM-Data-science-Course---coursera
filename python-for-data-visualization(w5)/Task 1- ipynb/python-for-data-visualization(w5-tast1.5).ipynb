```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace with your actual seasonality data)
seasonality = np.random.rand(50)

# Create a figure and axis for the scatter plot
fig, ax = plt.subplots(figsize=(8, 6))

# Create the scatter plot
scatter = ax.scatter(range(len(seasonality)), seasonality, c=seasonality, cmap='Blues', s=100, alpha=0.7, label='Seasonality')
ax.set_xlabel('Time Period')
ax.set_ylabel('Seasonality')
ax.set_title('Impact of Seasonality on Automobile Sales')
ax.legend()

# Create a colorbar axis and add the colorbar
cax = fig.add_axes([0.92, 0.15, 0.02, 0.7])  # [x, y, width, height] for the colorbar
cbar = plt.colorbar(scatter, cax=cax)
cbar.set_label('Seasonality')

plt.show()
```
