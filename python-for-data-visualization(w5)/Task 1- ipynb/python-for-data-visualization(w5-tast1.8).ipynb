```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
expenditure = [200000, 150000, 100000]  # Replace with the actual values for different vehicle types during recession

# Labels for the pie chart
labels = ['Type A', 'Type B', 'Type C']  # Replace with your vehicle type labels

# Explode a slice if required (e.g., 'Type A')
explode = (0.1, 0, 0)

# Create the pie chart
plt.pie(expenditure, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Advertising Expenditure for Vehicle Types during Recession Period')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
