```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with your own data)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015],
    'GDP_Recession': [12.5, 12.7, 12.2, 11.9, 12.0, 12.3],
    'GDP_NonRecession': [13.0, 13.5, 13.8, 13.7, 14.0, 14.2]
}
# Create a DataFrame from the data
df = pd.DataFrame(data)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns

# Subplot 1: GDP during Recession
axs[0].plot(df['Year'], df['GDP_Recession'], marker='o', color='b')
axs[0].set_title('GDP During Recession')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('GDP')
axs[0].grid(True)

# Subplot 2: GDP during Non-Recession
axs[1].plot(df['Year'], df['GDP_NonRecession'], marker='o', color='g')
axs[1].set_title('GDP During Non-Recession')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('GDP')
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
# Display the subplots
plt.show()
```
